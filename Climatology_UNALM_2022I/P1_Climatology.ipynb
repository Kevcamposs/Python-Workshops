## <font color = 'D9EC50'> 2. Ángulo de salida del sol

```python
import numpy as np
import matplotlib.pyplot as plt

nd = np.arange(0, 365, 1)  # Días
Decl = -23.45*np.cos(360/365*np.pi/180*(nd + 11))  # Declinación solar

L = -12.0  # Latitud: 12°S
H = np.arccos(-np.tan(L*np.pi/180)*np.tan(Decl*np.pi/180))*180/np.pi   # Ángulo de salida del sol

plt.plot(nd, H)
plt.title("Ángulo de salida del sol", fontsize = 24)
plt.grid()
plt.savefig("As.png")
```

```python
import numpy as np
import matplotlib.pyplot as plt

nd = np.arange(0, 365, 1)
Decl = -23.45*np.cos(360/365*np.pi/180*(nd + 11))

def H(L):
  return np.arccos(-np.tan(L*np.pi/180)*np.tan(Decl*np.pi/180))*180/np.pi

H0 = H(0)
H12 = H(-12)
H30 = H(-30)
H60 = H(-60)

plt.plot(nd, H0, label = "0°")
plt.plot(nd, H12, label = "12°S")
plt.plot(nd, H30, label = "30°S")
plt.plot(nd, H60, label = "60°S")

plt.title("Ángulo de salida del sol", fontsize = 24)
plt.legend()
plt.grid()
plt.savefig("As_L.png")
```

## <font color = '92EC50'> 3. Fotoperíodo - duración del día

```python
import numpy as np
import matplotlib.pyplot as plt

nd = np.arange(0, 365, 1)  # Días
Decl = -23.45*np.cos(360/365*np.pi/180*(nd + 11))  # Declinación solar

L = -12.0  # Latitud: 12°S
H = np.arccos(-np.tan(L*np.pi/180)*np.tan(Decl*np.pi/180))*180/np.pi   # Ángulo de salida del sol

N = 2*H/15   # Fotoperíodo

plt.plot(nd, N)
plt.title("Fotoperíodo", fontsize = 24)
plt.xlabel("Días")   # Nombre del eje X
plt.ylabel("Horas")  # Nombre del eje y
plt.grid()
plt.savefig("Fotoperíodo.png")
```

```python
import numpy as np
import matplotlib.pyplot as plt

nd = np.arange(0, 365, 1)
Decl = -23.45*np.cos(360/365*np.pi/180*(nd + 11))

def H(L):
  return np.arccos(-np.tan(L*np.pi/180)*np.tan(Decl*np.pi/180))*180/np.pi

N0 = 2*H(0)/15
N12 = 2*H(-12)/15
N30 = 2*H(-30)/15
N60 = 2*H(-60)/15

plt.plot(nd, N0, label = "0°")
plt.plot(nd, N12, label = "12°S")
plt.plot(nd, N30, label = "30°S")
plt.plot(nd, N60, label = "60°S")

plt.title("Fotoperíodo", fontsize = 24)
plt.legend()
plt.grid()
plt.savefig("N_L.png")
```

## <font color = '50ECD4'> 4. Hora de salida y puesta del sol

```python
import numpy as np
import matplotlib.pyplot as plt

nd = np.arange(0, 365, 1)
Decl = -23.45*np.cos(360/365*np.pi/180*(nd + 11))

def H(L):
  return np.arccos(-np.tan(L*np.pi/180)*np.tan(Decl*np.pi/180))*180/np.pi

fig, ax = plt.subplots(nrows = 1, ncols = 2,
                       figsize = (12, 6))

# Hora de salida del sol
Hss0 = 12 - H(0)/15
Hss12 = 12 - H(-12)/15
Hss30 = 12 - H(-30)/15
Hss60 = 12 - H(-60)/15

ax[0].plot(nd, Hss0, label = "0°")
ax[0].plot(nd, Hss12, label = "12°S")
ax[0].plot(nd, Hss30, label = "30°S")
ax[0].plot(nd, Hss60, label = "60°S")

ax[0].set_title("Hora de salida del sol", fontsize = 24, color = 'red')
ax[0].legend()
ax[0].grid()

# Hora de puesta del sol
Hss0 = 12 + H(0)/15
Hss12 = 12 + H(-12)/15
Hss30 = 12 + H(-30)/15
Hss60 = 12 + H(-60)/15

ax[1].plot(nd, Hss0, label = "0°")
ax[1].plot(nd, Hss12, label = "12°S")
ax[1].plot(nd, Hss30, label = "30°S")
ax[1].plot(nd, Hss60, label = "60°S")

ax[1].set_title("Hora de puesta del sol", fontsize = 24, color = 'blue')
ax[1].legend()
ax[1].grid()

plt.savefig("Hora.png")
```

## <font color = '7C65F4'> 5. Distancia Tiera-Sol

```python
c1 = 1.00011
c2 = 0.033523
c3 = 0.00128
c4 = 0.000739
c5 = 0.000099 
```

```python
import numpy as np
import matplotlib.pyplot as plt

nd = np.arange(0, 365, 1)
Theta = 2*np.pi*nd/365

Dts = c1 + c2*np.cos(Theta) + c3*np.sin(Theta) + c4*np.cos(2*Theta) + c5*np.cos(2*Theta)

fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize = (8, 6))

ax.plot(nd, Dts)

ax.set_title("Distancia Tierra-Sol", fontsize = 24, color = 'green')
ax.legend()
ax.grid()

plt.savefig("DistanciaTierraSol.png")
```

```python
import numpy as np
import matplotlib.pyplot as plt

nd = np.arange(0, 365, 1)

def Qs(L):
  L = L*np.pi/180
  nd = np.arange(0, 365, 1)
  Decl = -23.45*np.cos(360/365*np.pi/180*(nd + 11))
  H = np.arccos(-np.tan(L)*np.tan(Decl*np.pi/180))*180/np.pi

  Theta = 2*np.pi*nd/365
  c1 = 1.00011
  c2 = 0.033523
  c3 = 0.00128
  c4 = 0.000739
  c5 = 0.000099 
  Dts = c1 + c2*np.cos(Theta) + c3*np.sin(Theta) + c4*np.cos(2*Theta) + c5*np.cos(2*Theta)

  S = 1395 # W/m2
  Qs = S/np.pi*(H*np.pi/180*np.sin(L)*np.sin(Decl*np.pi/180) + np.cos(L)*np.cos(Decl*np.pi/180)*np.sin(H*np.pi/180))
  return Qs

fig, ax = plt.subplots(nrows = 1, ncols = 1, figsize = (8, 6))

# Radiación al tope de la atmósfera
Qs0 = Qs(0)
Qs30 = Qs(-30)
Qs45 = Qs(-45)
Qs60 = Qs(-60)

plt.plot(nd, Qs0, label = "0°")
plt.plot(nd, Qs30, label = "30°S")
plt.plot(nd, Qs45, label = "45°S")
plt.plot(nd, Qs60, label = "60°S")

ax.set_title("Radiación al tope de la atmósfera", fontsize = 24, color = 'darkblue')
ax.legend(fontsize = 12)
ax.grid()
ax.set_xlabel("Días", fontsize = 20)
ax.set_ylabel("W/m2", fontsize = 20)

plt.savefig("Qs_L.png")
```
